```python
import requests
from bs4 import BeautifulSoup
headers={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/95.0.4638.69 Safari/537.36'}
```

```python
cities=[]
city_links=[]
data = requests.get('https://www.swiggy.com/coimbatore')
soup=BeautifulSoup(data.text)
body=soup.find_all('a',{'class':'_15mJL'})
for item in body:
    cities.append(item.text)
    city_links.append(item.get('href'))
for i in range(0,len(city_links),1):
    city_links[i]=("https://www.swiggy.com"+city_links[i]+"?page={}")
print(cities)
print(city_links)
```

```python
hotel_cities=[]
links=[]
names=[]
ratings=[]
def getreport(city,page,no):
    data = requests.get(page.format(no))
    soup=BeautifulSoup(data.text)
    body=soup.find_all('div',{'class':'_3XX_A'})
    for item in body:
        hotel_cities.append(city)
        link=item.find('a',{'class':'_1j_Yo'})
        links.append(link.get('href'))
        name=item.find('div',{'class':'nA6kb'})
        names.append(name.text)
        rating=item.find_all('span')[1]
        ratings.append(rating.text)
```

```python
import pandas as pd
for i in range(0,len(cities),1):
    print(i)
    for j in range(1,10,1):
        getreport(cities[i],city_links[i],j)
for i in range(0,len(names),1):
    links[i]=("https://www.swiggy.com"+links[i])
```

```python
df={"Hotel":names,
    "Link":links,
    "City":hotel_cities,
    "Rating":ratings}
df=pd.DataFrame(df)
df=df.drop_duplicates(subset=["Hotel","Link","City"])
names=list(df["Hotel"])
links=list(df["Link"])
ratings=list(df["Rating"])
hotel_cities=list(df["City"])
```
